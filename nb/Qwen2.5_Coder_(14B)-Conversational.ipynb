We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, phi3, llama3` and more.

````python
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "qwen-2.5",
)

def formatting_prompts_func(examples):
    convos = examples["conversations"]
    texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
    return { "text" : texts, }
pass

from datasets import load_dataset
dataset = load_dataset("mlabonne/FineTome-100k", split = "train")
````

### Train the model
Now let's train our model. We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

The trainer includes our **gradient accumulation bug fix**. Read more about it here: [Blog post](https://unsloth.ai/blog/gradient)

````python
from trl import SFTConfig, SFTTrainer
from transformers import DataCollatorForSeq2Seq
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    data_collator = DataCollatorForSeq2Seq(tokenizer = tokenizer),
    packing = False, # Can make training 5x faster for short sequences.
    args = SFTConfig(
        per_device_train_batch_size = 1,
        gradient_accumulation_steps = 4, # Fixed major bug in latest Unsloth
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 30,
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "paged_adamw_8bit", # Save more memory
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use this for WandB etc
    ),
)
````

We also use Unsloth's `train_on_completions` method to only train on the assistant outputs and ignore the loss on the user's inputs.

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<|im_start|>user\n",
    response_part = "<|im_start|>assistant\n",
)
````

````python
# ---------- Cell A: Helpers, validate, parse tests, import patcher ----------
import ast, re, json, time, torch
from collections import defaultdict

# safe extract assistant text
def extract_assistant_text(decoded):
    parts = re.split(r"<\|im_start\|>assistant\s*\n", decoded, flags=re.IGNORECASE)
    if len(parts) >= 2:
        out = parts[-1]
        out = re.split(r"<\|im_end\|>", out)[0]
        return out.strip()
    return decoded.strip()

def extract_function_code(text):
    m = re.search(r"```(?:python)?\n(.+?)```", text, flags=re.S | re.I)
    if m:
        candidate = m.group(1).strip()
    else:
        candidate = text.strip()
    idx = candidate.find("def ")
    return candidate[idx:].strip() if idx != -1 else candidate

def parse_test_list_field(s):
    # parse list-like string into Python list of assert strings
    try:
        parsed = ast.literal_eval(s)
        if isinstance(parsed, (list, tuple)):
            return [str(x) for x in parsed]
        return [str(parsed)]
    except Exception:
        tokens = re.findall(r"assert\s.+?(?=(?:assert|$))", s, flags=re.S)
        return [t.strip() for t in tokens if t.strip()]

# Validate code by executing asserts in a temporary namespace
def validate_function(code_str, assert_list, auto_patch=True):
    ns = {}
    code_to_exec = code_str
    # Try to auto-insert imports if requested (basic patch)
    if auto_patch:
        imports = detect_needed_imports_from_error_hint(code_str)  # best-effort; function below defined
        if imports:
            # prepend imports unless already present
            for im in imports:
                if im not in code_to_exec:
                    code_to_exec = im + "\n" + code_to_exec
    try:
        exec(code_to_exec, ns)
    except Exception as e:
        # If NameError, optionally attempt simple import patching (common libs)
        err = repr(e)
        if "NameError" in err and auto_patch:
            patch_lines = auto_import_lines_from_error(err)
            if patch_lines:
                patched = patch_lines + "\n" + code_str
                try:
                    exec(patched, ns)
                    # continue to assert-check below with patched namespace
                    code_to_exec = patched
                except Exception as e2:
                    return False, "exec_error:" + repr(e2)
        else:
            return False, "exec_error:" + err
    # run asserts
    for a in assert_list:
        try:
            ast_node = a.strip()
            if ast_node.startswith("assert"):
                exec(ast_node, ns)
            else:
                val = eval(ast_node, ns)
                if not bool(val):
                    return False, "assertion_false:" + ast_node
        except AssertionError:
            return False, "assert_failed:" + a
        except Exception as e:
            return False, "assert_error:" + repr(e)
    return True, "ok"

# Small helper: derive import lines from NameError text
def auto_import_lines_from_error(errtxt):
    # errtxt like: "NameError(\"name 'Counter' is not defined\")"
    m = re.search(r"name '([A-Za-z0-9_]+)' is not defined", errtxt)
    if not m:
        return ""
    name = m.group(1)
    mapping = {
        "Counter": "from collections import Counter",
        "defaultdict": "from collections import defaultdict",
        "deque": "from collections import deque",
        "re": "import re",
        "math": "import math",
        "itertools": "import itertools",
        "heapq": "import heapq",
        "json": "import json",
        "datetime": "import datetime",
        "functools": "import functools",
    }
    return mapping.get(name, "")

# Another heuristic: detect certain keywords in code and suggest imports
def detect_needed_imports_from_error_hint(code_str):
    imports = []
    if "Counter" in code_str and "from collections" not in code_str:
        imports.append("from collections import Counter")
    if "re." in code_str and "import re" not in code_str:
        imports.append("import re")
    if "math." in code_str and "import math" not in code_str:
        imports.append("import math")
    return "\n".join(imports)

# Checkpoint utilities
from pathlib import Path
def save_checkpoint(results, fname="submission_partial.json"):
    with open(fname, "w", encoding="utf-8") as f:
        json.dump(results, f, ensure_ascii=False, indent=2)

def load_checkpoint(fname="submission_partial.json"):
    p = Path(fname)
    if p.exists():
        with open(fname, "r", encoding="utf-8") as f:
            return json.load(f)
    return []

print("Helpers loaded.")
````

````python
# ---------- Cell B: Build few-shot prompt prefix from trial.csv ----------
import pandas as pd
trial = pd.read_csv("trial.csv")
# Choose k examples; pick short ones to keep token use low
K = 2
examples = []
for _, row in trial.head(20).iterrows():  # small pool
    examples.append((row['instruction'], row.get('response', '')))
examples = examples[:K]

# Build few-shot text: instruction + response example
fewshot_prefix = ""
for inst, resp in examples:
    # shorten response to reasonable length (avoid huge tokens)
    resp_short = ("\n".join(resp.splitlines()[:20])) if isinstance(resp, str) else ""
    fewshot_prefix += f"Example:\nInstruction: {inst}\nOutput:\n{resp_short}\n\n"

print("Few-shot prefix built with", len(examples), "examples.")
````

````python
# ---------- Cell C: Dev tuning (try combos) ----------
import pandas as pd, time
from tqdm import tqdm

dev = pd.read_csv("dev_v2.csv")
# use a small dev subset for fast tuning
dev_subset = dev.head(20)

# hyperparam grid to test (small shortlist)
candidate_grid = [4, 6, 8]      # candidates to sample per item
temp_grid = [0.0, 0.15, 0.25]   # temperatures (0.0 disables sampling)
regen_grid = [1, 2]             # repair attempts

# generation function (returns list of candidate code strings)
def gen_candidates_one(inst, testlist, n, temp, max_new_tokens=300):
    messages = [{"role":"user","content": fewshot_prefix + inst + "\n\nVisible tests (do NOT print):\n" + str(testlist) + "\nReturn ONLY the function definition starting with 'def ' and include imports."}]
    inputs_tok = tokenizer.apply_chat_template(messages, tokenize=True, add_generation_prompt=True, return_tensors="pt")
    # move to device robustly
    if isinstance(inputs_tok, dict):
        for k,v in list(inputs_tok.items()):
            if hasattr(v, "to"): inputs_tok[k] = v.to(device)
        input_ids = inputs_tok["input_ids"]; attention_mask = inputs_tok.get("attention_mask", None)
    else:
        input_ids = inputs_tok.to(device); attention_mask=None

    with torch.no_grad():
        outputs = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            max_new_tokens=max_new_tokens,
            do_sample=(temp>0.0),
            temperature=temp,
            top_p=0.95,
            num_return_sequences=n if temp>0.0 else 1,
            use_cache=True,
        )
    decoded = tokenizer.batch_decode(outputs, skip_special_tokens=False)
    cands = [extract_function_code(extract_assistant_text(d)) for d in decoded]
    # unique
    seen=set(); uniq=[]
    for c in cands:
        if c not in seen:
            uniq.append(c); seen.add(c)
    return uniq

def run_dev_eval(n_candidates, temp, regen_tries):
    total=0; passed=0
    for _,row in dev_subset.iterrows():
        total += 1
        inst = str(row['instruction'])
        testlist = row['test_list']
        asserts = parse_test_list_field(testlist)
        # gen candidates
        cands = gen_candidates_one(inst, testlist, n_candidates, temp)
        ok=False
        for c in cands:
            v,info = validate_function(c, asserts, auto_patch=True)
            if v:
                ok=True; break
        # repair attempts if none passed
        attempts=0
        while not ok and attempts < regen_tries:
            # feed most promising candidate back for repair
            last = cands[0] if cands else ""
            repaired = repair_with_feedback_stub = None
            # simple repair: build a prompt with failure info; reuse gen_candidates_one with temp=0
            messages = [{"role":"user","content": fewshot_prefix + inst + "\n\nPrevious code:\n" + (last or "") + "\nFailure: " + (info if 'info' in locals() else 'failed') + "\nFix and return only corrected function."}]
            inputs_tok = tokenizer.apply_chat_template(messages, tokenize=True, add_generation_prompt=True, return_tensors="pt")
            if isinstance(inputs_tok, dict):
                for k,v in list(inputs_tok.items()):
                    if hasattr(v,"to"): inputs_tok[k] = v.to(device)
                input_ids = inputs_tok["input_ids"]; attn = inputs_tok.get("attention_mask", None)
            else:
                input_ids = inputs_tok.to(device); attn=None
            with torch.no_grad():
                out = model.generate(input_ids=input_ids, attention_mask=attn, max_new_tokens=300, do_sample=False, temperature=0.0, use_cache=True)
            dec = tokenizer.batch_decode(out, skip_special_tokens=False)
            repaired = extract_function_code(extract_assistant_text(dec[0]))
            cands = [repaired] + cands
            v, info = validate_function(repaired, asserts, auto_patch=True)
            if v:
                ok=True; break
            attempts += 1
        if ok:
            passed += 1
    return passed, total

# Run grid search (fast; this will take some time):
results_grid = []
for n in candidate_grid:
    for t in temp_grid:
        for r in regen_grid:
            print("Testing N=", n, "temp=", t, "regen=", r)
            start = time.time()
            p, tot = run_dev_eval(n, t, r)
            dur = time.time()-start
            acc = p/tot
            print(f"-> Acc {p}/{tot} = {acc:.4f} (took {dur:.1f}s)")
            results_grid.append({"n":n,"temp":t,"regen":r,"passed":p,"total":tot,"acc":acc,"time":dur})

# Print best few
results_grid = sorted(results_grid, key=lambda x:(-x['acc'], x['time']))
print("Top results:", results_grid[:5])
````

````python
# ---------- Cell D: Final full run with checkpointing ----------
import pandas as pd, json, time
from tqdm import tqdm
from pathlib import Path

CSV_PATH = "test_v2.csv"
SAVE_PARTIAL = "submission_partial.json"
FINAL_NAME = "submission.json"

N_CANDIDATES = 8      # set from dev tuning
TEMP = 0.2            # set from dev tuning
REGEN_TRIES = 2
MAX_NEW_TOKENS = 350
BATCH_SIZE = 1

df = pd.read_csv(CSV_PATH)
# resume partials
existing = load_checkpoint(SAVE_PARTIAL)
done_ids = {int(x['id']) for x in existing} if existing else set()
results = existing[:] if existing else []

items = [(int(r['id']), str(r['instruction']), str(r['test_list'])) for _,r in df.iterrows()]

for rid, inst, testlist in tqdm(items, desc="FinalRun"):
    if rid in done_ids:
        continue
    asserts = parse_test_list_field(testlist)
    # 1) generate N candidates (sampling)
    candidates = gen_n_candidates = None
    # generate candidates
    messages = [{"role":"user","content": fewshot_prefix + inst + "\n\nVisible tests (do NOT print):\n" + str(testlist) + "\nReturn ONLY the function definition starting with 'def ' and include imports."}]
    inputs_tok = tokenizer.apply_chat_template(messages, tokenize=True, add_generation_prompt=True, return_tensors="pt")
    if isinstance(inputs_tok, dict):
        for k,v in list(inputs_tok.items()):
            if hasattr(v,"to"): inputs_tok[k] = v.to(device)
        input_ids = inputs_tok["input_ids"]; attn = inputs_tok.get("attention_mask", None)
    else:
        input_ids = inputs_tok.to(device); attn=None
    with torch.no_grad():
        outputs = model.generate(
            input_ids=input_ids,
            attention_mask=attn,
            max_new_tokens=MAX_NEW_TOKENS,
            do_sample=(TEMP>0.0),
            temperature=TEMP,
            top_p=0.95,
            num_return_sequences=N_CANDIDATES if TEMP>0.0 else 1,
            use_cache=True,
        )
    decoded = tokenizer.batch_decode(outputs, skip_special_tokens=False)
    candidates = [extract_function_code(extract_assistant_text(d)) for d in decoded]
    # dedup
    seen=set(); uniq=[]
    for c in candidates:
        if c not in seen:
            uniq.append(c); seen.add(c)
    candidates = uniq
    # test candidates
    picked=None; last_err=None
    for cand in candidates:
        ok, info = validate_function(cand, asserts, auto_patch=True)
        if ok:
            picked=cand; break
        last_err=info
    # repair loop
    attempts=0
    while picked is None and attempts < REGEN_TRIES:
        last = candidates[0] if candidates else ""
        # produce repair (greedy)
        extra = f"\nPrevious code:\n{last}\nFailure: {last_err}\nFix and return only corrected function."
        messages = [{"role":"user","content": fewshot_prefix + inst + "\n\n" + extra}]
        inputs_tok = tokenizer.apply_chat_template(messages, tokenize=True, add_generation_prompt=True, return_tensors="pt")
        if isinstance(inputs_tok, dict):
            for k,v in list(inputs_tok.items()):
                if hasattr(v,"to"): inputs_tok[k] = v.to(device)
            input_ids = inputs_tok["input_ids"]; attn = inputs_tok.get("attention_mask", None)
        else:
            input_ids = inputs_tok.to(device); attn=None
        with torch.no_grad():
            out = model.generate(input_ids=input_ids, attention_mask=attn, max_new_tokens=MAX_NEW_TOKENS, do_sample=False, temperature=0.0)
        dec = tokenizer.batch_decode(out, skip_special_tokens=False)
        repaired = extract_function_code(extract_assistant_text(dec[0]))
        candidates = [repaired] + candidates
        ok, info = validate_function(repaired, asserts, auto_patch=True)
        if ok:
            picked = repaired; break
        last_err = info
        attempts += 1
        time.sleep(0.2)
    if picked is None:
        picked = candidates[0] if candidates else ""
    results.append({"id": int(rid), "response": picked})
    # save checkpoint each 20 items
    if len(results) % 20 == 0:
        save_checkpoint(results, SAVE_PARTIAL)

# final save
with open(FINAL_NAME, "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)
# remove partial if exists
p = Path(SAVE_PARTIAL)
if p.exists(): p.unlink()
print("Saved final", FINAL_NAME, "with", len(results), "rows.")
````
